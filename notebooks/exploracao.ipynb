### COLUNAS DATATIME

```python
colunas_data = ['data_primeira_compra', 'data_ultima_compra','data_pedido']
for coluna in colunas_data:
    df[coluna] = pd.to_datetime(df[coluna], errors='coerce',format='%Y-%m-%d')
df[['data_primeira_compra', 'data_ultima_compra','data_pedido']]
```

### COLUNAS FLOAT

```python
colunas_float = ['preco_unitario','despesas_operacionais', 'saldo_caixa', 'saldo_divida', 'gasto_mensal', 'percentual_desconto']
for coluna in colunas_float:
    df[coluna] = pd.to_numeric(df[coluna], errors='coerce')
    df[coluna] = df[coluna].replace({',': '.'}, regex=True).astype(float)
df[['preco_unitario','despesas_operacionais', 'saldo_caixa', 'saldo_divida', 'gasto_mensal', 'percentual_desconto']]
colunas_data = ['data_primeira_compra', 'data_ultima_compra','data_pedido']
for coluna in colunas_data:
    df[coluna] = pd.to_datetime(df[coluna], errors='coerce',format='%Y-%m-%d')
df[['data_primeira_compra', 'data_ultima_compra','data_pedido']]
```

### COLUNAS INTEIROS

```python
colunas_integer = ['quantidade']
for coluna in colunas_integer:
    df[coluna] = pd.to_numeric(df[coluna], errors='coerce').astype('Int64')
df[ ['quantidade']]
```

## COLUNAS TRATADAS

```python
colunas_int = ['quantidade']
colunas_float = ['preco_unitario','despesas_operacionais', 'saldo_caixa', 'saldo_divida', 'gasto_mensal', 'percentual_desconto']
colunas_data = ['data_primeira_compra', 'data_ultima_compra','data_pedido']

for coluna in colunas_int:
    df[coluna] = df[coluna].replace({' ': ''}, regex=True)
    df[coluna] = pd.to_numeric(df[coluna], errors='coerce').astype('Int64')

for coluna in colunas_float:
    df[coluna] = df[coluna].replace({',': '.'}, regex=True)
    df[coluna] = df[coluna].replace({' ': ''}, regex=True)
    df[coluna] = pd.to_numeric(df[coluna], errors='coerce').astype(float)

for coluna in colunas_data:
    df[colunas_data] = df[colunas_data].replace(r'\s+', '', regex=True)
    df[coluna] = pd.to_datetime(df[coluna], errors='coerce',format='%Y-%m-%d')
```
